```python
import pandas as pd

ren = pd.read_clipboard()
# ren.to_csv(r'assets/covid_renewable.csv')
ren.head()
```

```python
ren.drop(['(TWH)',	'SHARE',	'CROSS-ZONAL','(%)','SHARE.1', 'DOMESTIC', '(%).1'], axis=1, inplace=True)
ren.columns = ['year', 'total_traded_volume', 'cross_zonal_share', 'domestic_share']
ren.head()
```

```python
ren.cross_zonal_share = ren.cross_zonal_share.str.rstrip('%').astype(int)
ren.domestic_share = ren.domestic_share.str.rstrip('%').astype(int)
ren.head()
```

```python
ren['czs'] = ren.total_traded_volume * ren.cross_zonal_share / 100
ren['dms'] = ren.total_traded_volume * ren.domestic_share / 100
ren.head()
```

```python
ren.drop(['cross_zonal_share', 'domestic_share'], axis=1, inplace=True)
ren.to_csv(r'assets/electricity_trade.csv')
```
